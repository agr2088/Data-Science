### 1. Creating a list of five integers and printing the result of each operation: append, extend, insert, remove, pop, index, count, sort, reverse and clear.

```python
numbers = [50,40,30,20,10]
print(f"Original List: {numbers}")
```

```python
# Append used to add an element.
numbers.append("Alpha")
print(f"After Append{numbers}")
```

```python
# Extend used to add multiple elements.
numbers.extend(["Alpha","Alpha","Beta","Sigma"])
print(f"After Extend{numbers}")
```

```python
# Insert used to add an element at specific position.
numbers.insert(8,"Gamma") # inserting "Gamma" at index 8
print(f"After Insert{numbers}")
```

```python
# Count used to count the number of occurrences of a value in a list.
numbers.count("Alpha")
```

```python
# Sort used to sort elements in order.
numbers.sort(key=str) #key=str is used because my list contains strings.
print(numbers)
```

```python
# Reverse is used to reverse the sorted list.
numbers.reverse()
numbers
```

```python
# Index used to find an element.
numbers.index(50)
```

```python
# Remove used to remove an element.
numbers.remove("Gamma")
print(f"After Remove{numbers}")
```

```python
# Pop used to remove an element.
numbers.pop()     # removing last element
print(f"After Pop{numbers}")
```

```python
# Claer is used to empty the list.
numbers.clear()
numbers
```

### 6.	Given a 'marks' variable, print the grade:90 and above: 'A', 75–89: 'B' , 50–74: 'C' , Below 50: 'Fail'


```python
# Using marks to calulate the grade.(user_input)
marks = int(input("Enter the marks (0 ‑ 100): "))

if marks >= 90:                 # above 90
    grade = "A"
elif marks >= 75:              # 75‑89
    grade = "B"
elif marks >= 50:              # 50‑74
    grade = "C"
else:                          # anything below 50
    grade = "Fail"

print(f"Marks: {marks} → Grade: {grade}")
```

### 7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
num = int(input("Enter a number: "))

if num > 0:     # to get positive numbers.
                # inside a positive block we can also test for evenness
    if num % 2 == 0:
        print("Positive and even")
    else:
        print("Positive but odd")
elif num == 0:  # not positive -- is zero
    print("Zero")
else:           # negative numbers
    print("Negative")
```

### 10.	Create a 2D numpy array of shape (3x3). Convert it into a pandas DataFrame and add column names: 'A', 'B', 'C'.

```python
import numpy as np
import pandas as pd

# 3 × 3 NumPy array (here we just use 90‑99 for clarity)
a = np.arange(90,99).reshape(3, 3)      # shape = (3, 3)

# Converting into pandas DataFrame and adding the columns names as: A, B, C
df = pd.DataFrame(a, columns=['A', 'B', 'C'])
print(df)
```

### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:- Use .info(), .describe() , - Select 'Name' and 'City' columns , - Drop 'City' column ,- Fill any missing values in 'Salary' column with the mean , - Remove any duplicate rows.


```python
import numpy as np
import pandas as pd
# Creating a DataFrame with the given instruction.
df = pd.DataFrame({
    "Name"  : ["Alice","Bob","Carol","David","Eva","Frank","Grace","Hank","Ivy","Alice"],
    "Age"   : [23,34,29,42,31,27,38,45,26,23],
    "City"  : ["NY","Paris","Tokyo","Berlin","Sydney","London","Toronto","Dubai","Mumbai","NY"],
    "Salary": [56000, np.nan, 72000, 88000, 64000,48000, 73000, np.nan, 69000, 56000]})
df
```

```python
df.info() # Checking the information of DataFrame.
```

```python
df.describe() # Checking the description of DataFrame.
```

```python
df[["Name","City"]] # to get the columns "Name" and "City".
```

```python
df =df.drop(columns=["City"]) # To remove the column "City".
df
```

```python
mean_salary = df["Salary"].mean() # Calculating the mean of Salary for missing values.
df["Salary"] = df["Salary"].fillna(mean_salary)
df
```

```python
df = df.drop_duplicates().reset_index(drop=True) # Removing duplicate rows and resetting index.
df
```
