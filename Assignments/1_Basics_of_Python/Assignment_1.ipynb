## Exercise 1: Prime Numbers

```python
# Finding a prime number using for loop and if conditions along with user input.

number = int(input())
for i in range(2, number):
    if number % i == 0:
        print(f"{number} is not a prime number")
        break
else:
    print(f"{number} is a prime number")
```

## Exercise 2: Product of Random Numbers

```python
 #Generating two random numbers and verifing the product id and diplaying appropiate message.

import random
a,b = random.randint(1, 12), random.randint(1, 12)
PRODUCT_ID = a * b
user_answer = int(input(f"Please Enter the product of {a} and {b}? "))

if user_answer == PRODUCT_ID:
    print("PRODUCT VERIFICATION SUCCESSFUL")
else:
    print("INVALID. PLEASE VERIFY AGAIN ")
```

## Exercise 3: Squares of Even/Odd Numbers

```python
# Using range function and if conditions to display the square of even & odd numbers.
for i in range(100, 201):
    print(f"{i} squared is {i ** 2} ({'even' if i % 2 == 0 else 'odd'})") 

# Using conditional statement to only include squared numbers that are even only.
for i in range(100, 201):
    if i % 2 == 0:
        print(f"{i} squared is {i ** 2}, even")
```

## Exercise 4:

```python
# Caluclating the final hospital bill & applying the discount if the total bill exceeds ₹10,000.

room_charge = 2000 * 4
doctor_fee = 1500
lab_tests = 300 * 3
medicine_cost = 2400

total_bill_before_discount = room_charge + doctor_fee + lab_tests + medicine_cost
discount = total_bill_before_discount * 0.10 if total_bill_before_discount > 10000 else 0
final_bill = total_bill_before_discount - discount

print(f"Total bill: ₹{room_charge} + ₹{doctor_fee} + ₹{lab_tests} + ₹{medicine_cost}")
print(f"Discount: ₹{(total_bill_before_discount * 0.10)}")
print(f"Final Bill: ₹{final_bill:.2f}")
```
